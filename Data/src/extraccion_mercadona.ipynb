Extraer el título de la marca del producto

```python
import re
food_label = []
for title in food_title_list:
    words = re.findall(r'[A-ZÁÉÍÓÚÜÑ][a-záéíóúüñ\s]*', title)
    if len(words) >= 2:
        food_label.append(' '.join(words[1:]))
    else:
        food_label.append('')
food_label

```

Obtener el nombre de los productos sin la marca

```python
productos_sin_marcas = []
for producto, marca in zip(food_title_list, marcas_productos):
    producto_sin_marca = producto.replace(marca, '').strip()
    productos_sin_marcas.append(producto_sin_marca)
print(productos_sin_marcas)
```

Extraer formato del producto

```python
food_format = driver.find_elements(By.CLASS_NAME, 'product-format.product-format__size--cell')
food_format_list = [e.text for e in food_format]
food_format_list
```

```python
len(food_format_list)
```

Se divide el formato en solo formato y cantidad

```python
import re
cantidad = []
solo_formato = []

for producto in food_format_list:
    match = re.search(r'\d+\D+.*', producto)
    if match:
        cantidad.append(match.group())
        solo_formato.append(re.sub(r'\d+\D+.*', '', producto).strip())
    else:
        solo_formato.append(producto.strip())

cantidad
solo_formato
```

```python
cantidad
```

Buscamos el precio de los productos

```python
price = driver.find_elements(By.CLASS_NAME, 'product-price')
precio_list = [e.text for e in price]
precio_list
```

Dividimos el precio y el lote del producto, el lote no lo necesitaremos para nada

Aquí se depura el precio final para quitarnos trabajo de limpieza

```python
import re


precios_finales = []
for precio in precio_list:
    precios_encontrados = re.findall(r'(\d+,\d+ €)(?=\s*/)', precio)
    precio_final = precios_encontrados[-1] if precios_encontrados else None
    precios_finales.append(precio_final)
print(precios_finales)
```
