```python
def categorizar_producto_final(producto):
    producto = producto.lower()
    if any(fruit in producto for fruit in ["manzana", "pera", "tomate", "zanahoria"]):
        return "Fruta y verdura"
    elif any(meat in producto for meat in ["pollo", "cerdo", "ternera", "carne"]):
        return "Carne"
    elif any(seafood in producto for seafood in ["pescado", "marisco", "salmón"]):
        return "Marisco y pescado"
    elif any(drink in producto for drink in ["agua", "refresco", "zumo"]):
        return "Agua y refrescos"
    elif any(bodega in producto for bodega in ["brandy", "vino", "cerveza", "whisky", "bebida espirituosa"]):
        return "Bodega"
    elif any(lacteo in producto for lacteo in ["leche", "yogur", "queso", "mantequilla", "huevo"]):
        return "Huevos, leche y mantequillla"
    else:
        return "Otro"


productos['categoría'] = productos['producto'].apply(categorizar_producto_final)
```

```python
productos = productos.drop('categoría', axis=1)
```
