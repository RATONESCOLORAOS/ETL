```python
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings

# Configuración de las opciones del driver
opciones = Options()
opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
opciones.add_experimental_option('useAutomationExtension', False)
opciones.headless = False  # True si no quieres que aparezca la ventana del navegador
opciones.add_argument('--start-maximized')  # Comienza maximizado
opciones.add_argument('--incognito')  # Modo incógnito

# Instalación del Chrome Driver y apertura del navegador con las opciones configuradas
PATH = ChromeDriverManager().install()  # Instala el driver de Chrome
driver = webdriver.Chrome(PATH, options=opciones)  # Asegúrate de pasar las opciones aquí

# Ignora las advertencias no relevantes para este script
warnings.filterwarnings('ignore')

# Navegar a la página
driver.get('https://www.dia.es/patatas-fritas-encurtidos-y-frutos-secos/aceitunas-y-encurtidos/c/L2096')
```

```python
import time
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    driver.execute_script("window.scrollBy(0, 1000)")
    time.sleep(2)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
food_title = driver.find_elements(By.CLASS_NAME, 'search-product-card__product-name')
food_title_list = [e.text for e in food_title]
print(food_title_list)
```

```python
len(food_title_list)
```

```python
def extraer_marca(producto, marcas):
    for marca in marcas:
        if marca in producto:
            return marca
    return "Marca no encontrada"
```

```python
marcas_productos = [extraer_marca(producto, product_title_list) for producto in food_title_list]

marcas_productos
```

```python

def eliminar_marca_producto(producto, marcas):
    for marca in marcas:
        if marca in producto:
            return producto.replace(marca, "").strip()
    return producto
```

```python

productos_sin_marca = [eliminar_marca_producto(producto, product_title_list) for producto in food_title_list]

productos_sin_marca
```

```python
product_format = ['lata','spray' ,'tarrina' ,'en rulo' ,'vaso' ,'brik' ,'botella','blister','bote', 'frasco', 'caja','bolsa','paquete', 'sobre', 'estuche', 'bandeja', 'rectangular' ]
```

```python
def extraer_formato(producto, formatos):
    for formato in formatos:
        if formato in producto:
            return formato
    return "Formato no encontrado"
```

```python
formatos_productos = [extraer_formato(producto, product_format) for producto in productos_sin_marca
]

formatos_productos
```

```python
len(formatos_productos)
```

```python
def eliminar_formato_producto(producto, formatos):
    for formato in formatos:
        if formato in producto:
            return producto.replace(formato, "").strip()
    return producto
```

```python
productos_sin_formato = [eliminar_formato_producto(producto, product_format) for producto in productos_sin_marca
]

productos_sin_formato
```

```python
import re
cantidad = []
solo_productos = []

for producto in productos_sin_formato:
    match = re.search(r'\d+\D+.*', producto)
    if match:
        cantidad.append(match.group())
        solo_productos.append(re.sub(r'\d+\D+.*', '', producto).strip())
    else:
        solo_productos.append(producto.strip())

cantidad
solo_productos
```

```python
import pandas as pd
data = {'producto': solo_productos,'marca': marcas_productos, 'formato': formatos_productos , 'cantidad': cantidad , 'precio': precio, 'supermercado': 'Dia'}
df = pd.DataFrame(data)
df
```
